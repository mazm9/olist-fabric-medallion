```python
from pyspark.sql.functions import col, sum as _sum, countDistinct, avg, month, year

sales = spark.table("silver_sales")

# KPIs
gold_kpi = (
    sales.groupBy(
        year("order_purchase_timestamp").alias("year"),
        month("order_purchase_timestamp").alias("month")
    )
    .agg(
        _sum("payment_value").alias("revenue"),
        countDistinct("order_id").alias("orders"),
        avg("payment_value").alias("avg_order_value")
    )
)

(
    gold_kpi.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("gold_sales_kpi")
)
```
